```python
import warnings

import sys
from pathlib import Path

# PROJECT_ROOT = Path().resolve().parents[1]
# sys.path.append(str(PROJECT_ROOT))


# Подавляем только FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

import pandas as pd

from train.inference_func import inference_batch
from train.ppo_volume import PPOWithImitation

pd.set_option("display.precision", 2)
pd.options.display.float_format = "{:,.2f}".format
```

```python
eth_with_ta = pd.read_csv("train/data_with_ta.csv", index_col="timestamp", parse_dates=True)

split_date_1 = pd.to_datetime("2025-01-01")
split_date_2 = pd.to_datetime("2025-02-01")

df_test = eth_with_ta[
    (eth_with_ta.index >= split_date_1) & (eth_with_ta.index < split_date_2)
].copy()

window_size = 96
excluded = ["future_return", "target", "target_expert"]

df_test = df_test.drop(columns=[col for col in excluded if col in df_test.columns])

df_test["future_return"] = df_test["close"].shift(-window_size) / df_test["close"] - 1
df_test["future_return"] = df_test["future_return"].fillna(0)

state_columns = [col for col in df_test.columns]

print(f"Размер df_test: {df_test.shape}")
print(f"Признаков для state: {len(state_columns)}")
```
